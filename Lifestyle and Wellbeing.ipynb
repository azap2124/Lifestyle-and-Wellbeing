### 4.3 Mindfulness and Meditation Insights

In this section, I restructured the database into two distinct groups:

- df_meditation: This data frame comprises individuals who reported engaging in weekly meditation more than 7 times.
- df_no_meditation: Within this data frame are individuals who reported practicing weekly meditation 6 times or less.

```python
# creating two separate data frames based on weekly meditation frequency

df_meditation = df[df['weekly_meditation'] >= 7]

df_no_meditation = df[df['weekly_meditation'] == 0]
```

I then wanted to compare the average stress levels between individuals actively engaging in meditation and those not practicing meditation. To accomplish this, I constructed a new data frame, df_meditation_avg, utilizing calculations from the meditation_avg variable, which computes the average stress levels for regular meditators and non-meditators.

As anticipated, the analysis revealed that individuals who meditate regularly experience significantly lower levels of stress compared to those who do not engage in regular meditation.

```python
# creating a new dataframe capturing the average daily stress based on meditation habits

meditation_avg = {'Meditates regularly' : df_meditation['daily_stress'].mean(),
                'Does not meditate regularly' : df_no_meditation['daily_stress'].mean()
                 }

# new dataframe 

df_meditation_avg = pd.DataFrame(list(meditation_avg.items()), columns=['Meditation Habit', 'Average Daily Stress'])

print(df_meditation_avg)
```

```python
# creating a bar chart representing the average daily stress levels based on meditation habits

sns.set(style="white")

plt.figure(figsize=(10, 6))

bars = plt.bar(df_meditation_avg['Meditation Habit'], df_meditation_avg['Average Daily Stress'], color='skyblue')

plt.bar_label(bars, fmt='%.2f', label_type='edge', fontsize=10, color='black')

plt.xlabel('Meditation Habit', fontsize=12)
plt.ylabel('Average Daily Stress', fontsize=12)
plt.ylim(0, 5)
plt.title('Average Daily Stress by Meditation Habit', fontsize=14)

plt.show()
```

#### Do individuals who engage in weekly meditation report higher levels of flow?

Next, I was interested in finding out if individuals who meditated regularly reported higher levels of flow. "Flow" occurs when an individual is fully immersed in a task, often losing track of time and experiencing a sense of energized focus, enjoyment, and engagement. The concept of flow and meditation are related in several ways. They both involve a heightened state of attention and being in the present moment so I'm interested how they relate in this dataset. 

To start, **I calculated the average flow levels of individuals who reported meditating more than 7 times per week and those who did not meditate at all.**

```python
# average flow levels of individuals who meditated regularly

mean_flow_meditation = df[df['weekly_meditation'] >= 7]['flow'].mean()

print(mean_flow_meditation)
```

```python
# average flow levels of individuals who did not meditated regularly

mean_flow_no_meditation = df[df['weekly_meditation'] == 0]['flow'].mean()

print(mean_flow_no_meditation)
```

```python
if mean_flow_meditation > mean_flow_no_meditation:
    print("Individuals who engage in weekly meditation report higher levels of flow.")
else:
    print("There is no significant difference in reported levels of flow between individuals who engage in weekly meditation and those who do not.")
```

```python
categories = ['Meditates Regularly', 'Does not Meditates Regularly']
mean_flow_values = [mean_flow_meditation, mean_flow_no_meditation]

sns.set(style="white")

plt.figure(figsize=(10,6))

flow_mean_bar = plt.bar(categories, mean_flow_values, color='skyblue')
plt.bar_label(flow_mean_bar, fmt='%.2f', label_type='edge', fontsize=10, color='black')

plt.ylabel('Average Flow')
plt.xlabel('Meditation Habits')
plt.title('Mean Flow for Different Weekly Meditation Durations')
plt.ylim(0,5)
plt.show()
```

###  4.4 Financial Well-being Patterns

The way individuals manage the balance between their work and personal lives plays a pivotal role in determining the amount of stress they encounter daily. To test this hypothesis, **I was interested in finding out how the Work-Life score compared to the daily stress of individuals.** To do this, I divided the data into two groups (as a reminder, "sufficient_income" is how an individual's income covered their basic life expenses):  

1) Individuals who had sufficient income (df_income)
2) Individuals who did not have sufficient income (df_no_income) 

Not surprisingly, individuals with insufficient income reported higher levels of stress and a lower Work-Life Balance score by a significant amount. 

```python
df_income = df[df['sufficient_income'] == 2]
df_no_income = df[df['sufficient_income'] == 1]

df_sufficient_wlb = df_income['work_life_balance_score'].mean()
df_insufficient_wlb = df_no_income['work_life_balance_score'].mean()

data = {'Income Status': ['Sufficient Income', 'Insufficient Income'],
        'Average Stress Levels': [df_income['daily_stress'].mean(), df_no_income['daily_stress'].mean()],
        'Work-Life Score': [df_sufficient_wlb, df_insufficient_wlb],
}

df_sufficient_income = pd.DataFrame(data)

print(df_sufficient_income)
```

```python
# bar plot for average stress levels

sns.set(style="white")

plt.figure(figsize=(10,6))

income_status_bar = sns.barplot(x='Income Status', y='Average Stress Levels', data=df_sufficient_income, color='skyblue')

income_status_bar.bar_label(income_status_bar.containers[0], fmt='%.2f', label_type='edge', fontsize=10, color='black')

plt.title('Average Stress Levels by Income Status')
plt.ylim(0, 4)

plt.show()
```

#### Do individuals with insufficient income encounter a less favorable work-life balance?

```python
# bar plot for work-life score

sns.set(style="white")

plt.figure(figsize=(10,6))

work_life_bar = sns.barplot(x='Income Status', y='Work-Life Score', data=df_sufficient_income, color='skyblue')

work_life_bar.bar_label(work_life_bar.containers[0], fmt='%.2f', label_type='edge', fontsize=10, color='black')

plt.title('Work-Life Score by Income Status')
plt.ylim(0,800)
plt.show()
```

### 4.7 What is the main feature of a happy life?
